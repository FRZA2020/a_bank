```python
import numpy as np

class bank():
     
    # init method or constructor
    def __init__(self):
        self.customers = [] # initialize a list for the customer names
        self.balances = np.array([]) # initialize a numpy array for the balance of each customer
        self.credit = np.array([])
    # add a customer, initialize deposit value with 0, if no input value is given
    
    def add_customer(self, customer_name, initial_balance = 0):
        if customer_name in self.customers: # if customer name already exists, don't add customer
            print('Customer', customer_name, 'already exists. Please choose a different/unique name.')
        else:
            self.customers.append(customer_name)
            self.balances = np.append(self.balances, initial_balance)
            self.credit = np.append(self.credit, 0)
            print('New customer', customer_name, 'created with', initial_balance, 'in deposit.')
        
    def deposit_withdraw(self, customer_name, amount, type_of_account, account):
        if not customer_name in self.customers: # checks whether customer is in bank
            print('Customer with name', customer_name, 'does not exist.')
        else:
            i = self.customers.index(customer_name) # get the index for the customer name from the list
            
            this_balance = account[i]
                #this_balance = self.balances[i] # pass index to numpy array to access balance for customer
            if amount < 0:
                if -amount > this_balance: # don't withdraw money if it results in a negative balance
                    print(customer_name, 'can withdraw no more than', this_balance)
                else:
                    print(customer_name, 'is withdrawing', -amount, 'from', type_of_account)
                    account[i] = account[i] + amount
            else:
                account[i] = account[i] + amount
                print(customer_name, 'has deposited', amount, 'in', type_of_account)
  
                    
    def show_customer_balance(self, customer_name, type_of_account, account):
        
            if not customer_name in self.customers: # check if customer_name exists in list of names for the bank
                    print('Customer with name', customer_name, 'does not exist.')
            else:
                i = self.customers.index(customer_name) # get the index for the customer name
                print('Balance of customer',customer_name, 'in', type_of_account, customer_name, ':',account[i]) # pass index to numpy array to access balance for customer
        
                
    def show_bank_balance(self):
        if not self.customers: # if not operator checks if the list self.customers is an empty list
            print('There are not yet any customers in the bank. ')
        else:
            # the sum of all entries in the numpy array for the balances is the total amount in the bank
            print('The total amount in the bank is:', np.sum(self.balances)) 
            
            

```

```python
mybank3 = bank()
mybank3.add_customer('Hugo', 10)
mybank3.show_customer_balance('Hugo', 'savings', mybank3.balances)
mybank3.show_customer_balance('Hugo', 'credit', mybank3.credit)
mybank3.deposit_withdraw('Hugo', 10, 'savings',mybank3.balances)
mybank3.show_customer_balance('Hugo', 'savings', mybank3.balances)
mybank3.show_customer_balance('Hugo', 'credit', mybank3.credit)
# deposit from credit account
mybank3.deposit_withdraw('Hugo', 15, 'credit',mybank3.credit)
mybank3.show_customer_balance('Hugo', 'savings', mybank3.balances)
mybank3.show_customer_balance('Hugo', 'credit', mybank3.credit)
# withdraw from credit account
mybank3.deposit_withdraw('Hugo', -25, 'credit',mybank3.credit)
mybank3.show_customer_balance('Hugo', 'savings', mybank3.balances)
mybank3.show_customer_balance('Hugo', 'credit', mybank3.credit)
mybank3.deposit_withdraw('Hugo', -2, 'credit',mybank3.credit)
mybank3.show_customer_balance('Hugo', 'savings', mybank3.balances)
mybank3.show_customer_balance('Hugo', 'credit', mybank3.credit)
```
